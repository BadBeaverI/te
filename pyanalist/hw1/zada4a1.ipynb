## На складе лежат разные фрукты в разном количестве.
### Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

```python
def total_fruits(*args):
    total = 0
    for fruit, quantity in args:
        total += quantity
    return total
fruits_on_stock = total_fruits(('яблоки', 10), ('бананы', 55), ('апельсины', 8))
print(f"Общее количество фруктов на складе: {fruits_on_stock}")
```

##  *Условие 3: *
### Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
## [100, 125, -90, 345, 655, -1, 0, 200]
# Используйте list comprehensions

```python
ad_expenses = [100, 125, -90, 345, 655, -1, 0, 200]

filtered_expenses = [expense for expense in ad_expenses if expense >= 0]

total_expenses = sum(filtered_expenses)

print("Список без отрицательных значений:", filtered_expenses)
print("Суммарные затраты:", total_expenses)
```
